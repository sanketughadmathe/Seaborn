# Distribution Plots


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 30)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

# Categorical Plots

```python
fig = plt.subplots(figsize = (8,6))
sns.barplot(x = 'sex', y ='total_bill', data = tips)
```

```python
fig = plt.subplots(figsize = (8,6))
sns.countplot(x = 'sex', data = tips)
```

```python
fig = plt.subplots(figsize = (10,6))
sns.boxplot(x = 'day',  y = 'total_bill', data = tips,palette='rainbow')
```

```python
fig = plt.subplots(figsize = (10,6))
sns.boxplot(x = 'day',  y = 'total_bill', data = tips, hue= 'smoker', palette="coolwarm")
```

```python
fig = plt.subplots(figsize = (10,6))
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
fig = plt.subplots(figsize = (10,6))
sns.violinplot(x = 'day',  y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
fig = plt.subplots(figsize = (10,6))
sns.stripplot(x = 'day',  y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.swarmplot(x = 'day',  y = 'total_bill', data = tips)
```

```python
fig = plt.subplots(figsize = (10,6))
sns.violinplot(x = 'day',  y = 'total_bill', data = tips)
sns.swarmplot(x = 'day',  y = 'total_bill', data = tips, color ='black')
```

```python
sns.catplot(x = 'day',  y = 'total_bill', data = tips, kind = 'bar')
```

 # Matrix Plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True)
```

```python
fp = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
fp
```

```python
fig= plt.subplots(figsize=(10,8))
sns.heatmap(fp, cmap ='coolwarm' , linecolor='black', linewidths = 2)
```

```python
sns.clustermap(fp, cmap = 'coolwarm', standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species', palette = 'prism')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```
